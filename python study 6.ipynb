```python
# Bubble sort
def bubble_sort(arr):
    n = len(arr)

    for i in range(n):
        swapped = False

        for j in range(0, n - i -1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True

        if not swapped:
            break

    return arr
```

```python
def quicksort(arr, low=0, high=None):
    if high is None:
        high = len(arr) - 1

    if low < high:
        pivot_index = partition(arr, low, high)

        quicksort(arr, low, pivot_index - 1)
        quicksort(arr, pivot_index + 1, high)

    return arr

def partition(arr, low, high):
    pivot = arr[high]

    i = low - 1

    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

def partition_hoare(arr, low, high):
    pivot = arr[low]
    i = low - 1
    j = high + 1

    while True:
        i += 1
        while arr[i] < pivot:
            i += 1

        j -= 1
        while arr[j] > pivot:
            j -= 1

        if i >= j:
            return j

        arr[i], arr[j] = arr[j], arr[i]

import random

def randomize_quicksort(arr, low=0, high=None):
    if high is None:
        high = len(arr) - 1

    if low < high:
        random_index = random.randint(low, high)
        arr[random_index], arr[high] = arr[high], arr[random_index]

        pivot_index = partition(arr, low, high)
        randomized_quicksort(arr, low, pivot_index -1)
        randomized_quicksort(arr, pivot_index + 1, high)

    return arr


def quicksort_visual(arr, low = 0, high=None, depth=0):
    if high is None:
        high = len(arr) - 1

    indent = "  " * depth
    print(f"{indent}Sorting {arr[low:high+1]}")

    if low < high:
        pivot_index = partition(arr, low, high)
        print(f"{indent}Pivot {arr[pivot_index]} at index {pivot_index}")
        print(f"{indent}Left: {arr[low:pivot_index]}, Right: {arr[pivot_index + 1: high + 1]}")
        
        quicksort_visual(arr, low, pivot_index - 1, depth + 1)
        quicksort_visual(arr, pivot_index + 1, high, depth + 1)

test_array = [64, 34, 25, 12, 22, 11, 90]
quicksort_visual(test_array.copy())
        
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    left_sorted = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    return merge(left_sorted, right_sorted)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result

def merge_sort_inplace(arr, left=0, right=None):
    if right is None:
        right = len(arr) - 1

    if left < right:
        mid = (left + right) // 2

        merge_sort_inplace(arr, left, mid)
        merge_sort_inplace(arr, mid + 1, right)

        merge_inplace(arr, left, mid, right)
    return arr

def merge_inplace(arr, left, mid, right):
    left_arr = arr[left: mid + 1]
    right_arr = arr[mid + 1: right + 1]

    i = j = 0
    k = left

    while i < len(left_arr) and j < len(right_arr):
        if left_arr[i] <= right_arr[j]:
            arr[k] = left_arr[i]
            i += 1
        else:
            arr[k] = right_arr[j]
            j += 1
        k += 1

    while i < len(left_arr):
        arr[k] = right_arr[j]
        j += 1
        k += 1

def merge_sort_visual(arr, depth=0):
    indent = "  " * depth
    print(f"{indent}Dividing: {arr}")

    if len(arr) <= 1:
        print(f"{indent}Base case: {arr}")
        return arr

    mid = len(arr) // 2
    left_half = arr[:mid]
    right_half = arr[mid:]

    print(f"{indent}Left: {left_half}, Right: {right_half}")

    left_sorted = merge_sort_visual(left_half, depth + 1)
    right_sorted = merge_sort_visual(right_half, depth + 1)

    merged = merge(left_sorted, right_sorted)
    print(f"{indent}Merged: {merged}")

    return merged

test_array = [64, 34, 25, 12, 22, 11, 90]
merge_sort_visual(test_array.copy())
    

```

```python
# HeapSort
def heap_sort(arr):
    n = len(arr)

    for i in range(n // 2-1, -1, -1):
        heapify(arr, n, i)

    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]

        heapify(arr, i, 0)

    return arr

def heapify(arr, n, i):
    largest = i 
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)

def heap_sort_visual(arr):
    print(f"Initial array: {arr}")
    n = len(arr)

    print("\nBuilding max heap:")
    for i in range(n // 2 - 1, -1, -1):
        heapify_visual(arr, n, i)
        print(f"After heapifying index {i}: {arr}")

    print(f"\nMax heap built: {arr}")

    print("\nExtracting elements:")
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        print(f"Moved {arr[i]} to position {i}: {arr}")

        heapify_visual(arr, i, 0)
        print(f"After heapifying: {arr[:i]} | {arr[i:]}")

    return arr

def heapify_visual(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify_visual(arr, n, largest)


test_array = [64, 34, 25, 12, 22, 11, 90]
heap_sort_visual(test_array.copy())

        
    
    
        
```
